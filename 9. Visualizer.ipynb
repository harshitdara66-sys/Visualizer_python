# 3. Low Level Data Understanding 

```python
# a. Find count of unique values in location column

Unique = DF['location'].unique()
print(f"The Unique values in the Location Column is given follow : \n\n{Unique}")
```

```python
# b. Find which continent has maximum frequency using values counts

Max_Continent = DF['continent'].value_counts()
Continent = Max_Continent.idxmax()
Frequency = Max_Continent.max()

print(f"The most frequently Continent is : {Continent} , Number of frequency : {Frequency}")
```

```python
# c. Find maximum & mean value in 'total_cases'

Max_Cases = DF['total_cases'].max()
Average_Cases = DF['total_cases'].mean().round(2)

print(f"The Maximum Number of Total Cases is : {Max_Cases}")
print(f"The Average Number of Total Cases is : {Average_Cases}")
```

```python
# d. Find 25%,50% & 75% quartile value in 'total_deaths'

Quartile_Total_Deaths = DF['total_deaths'].describe()
Quartile_Total_Deaths
```

```python
# e. Find which continent has maximum 'human_development_index'.

Continents = DF.groupby('continent')['human_development_index'].sum().sort_values(ascending = False)
Name_Continents = Continents.idxmax()
Total = Continents.max()

print(f"The Continent : {Name_Continents} who has maximum Human Development : {Total}")
```

```python
# f. Find which continent has minimum 'gdp_per_capita'.

Continents = DF.groupby('continent')['gdp_per_capita'].min().sort_values(ascending = True)
Name_Continents = Continents.idxmin()
Min_gdp_per_capita = Continents.min()

print(f"The Continent : {Name_Continents} who has Minimum GDP Per Capita : {Min_gdp_per_capita}")
```

# 5. Data Cleaning

```python
# a. Remove all duplicates observations

Duplicated_values = DF.duplicated().sum()
print(f"The number of Duplicated values in the Dataset is : {Duplicated_values}")
```

```python
# b. Find missing values in all columns

DF.isnull().sum()
```

```python
# c. Remove all observations where continent column value is missing

DF = DF.dropna(subset = ['continent'])
```

```python
# d. Fill all missing values with 0

DF.fillna(0 , inplace = True)
```

# 6. Date time format

```python
# a. Convert date column in datetime format using pandas.to_datetime

DF['date'] = pd.to_datetime(DF['date']).dt.date
DF['date']
```

```python
# b. Create new column month after extracting month data from date column

DF['Month'] = pd.to_datetime(DF['date']).dt.month
DF['Month'].sort_values(ascending = True)
```

# 7. Data Aggregation

```python
# a. Find max value in all columns using groupby function on 'continent' column
# b. Store the result in a new dataframe named 'df_groupby' (Use df_groupby dataframe for all further analysis)

df_groupby = DF.groupby('continent').max().reset_index()
df_groupby
```

# 8. Feature Engineering

```python
# a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
df_groupby['total_deaths_to_total_cases']
```

# 9. Data Visualization

```python
# a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

sns.distplot(df_groupby['gdp_per_capita'] , kde = True)
plt.title('Distribution of GDP per Capita')
plt.show()
```

```python
# b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

plt.scatter(x = df_groupby['total_cases'] , y = df_groupby['gdp_per_capita'] , color = 'darkgreen')
```

```python
# c. Plot Pairplot on df_groupby dataset.

sns.pairplot(df_groupby.select_dtypes(include='number'))
plt.show()
```
